```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
```

# Data Loading

```python
#loading the csv files and storing them in variables

pl10=pd.read_csv('cleaned/pl10.csv')
pl11=pd.read_csv('cleaned/pl11.csv')
pl12=pd.read_csv('cleaned/pl12.csv')
pl13=pd.read_csv('cleaned/pl13.csv')
pl14=pd.read_csv('cleaned/pl14.csv')
pl15=pd.read_csv('cleaned/pl15.csv')
pl16=pd.read_csv('cleaned/pl16.csv')
pl17=pd.read_csv('cleaned/pl17.csv')
pl18=pd.read_csv('cleaned/pl18.csv')
pl19=pd.read_csv('cleaned/pl19.csv')
pl20=pd.read_csv('cleaned/pl20.csv')
pl21=pd.read_csv('cleaned/pl21.csv')
```

# Data Wrangling

```python
pl10=pl10[(pl10['Name of the Organization']!='GRAND TOTAL')]
pl11=pl11[(pl11['Name of the Organization']!='GRAND TOTAL')]
pl12=pl12[(pl12['Name of the Organization']!='GRAND TOTAL')]
pl13=pl13[(pl13['Name of the Organization']!='GRAND TOTAL')]
pl14=pl14[(pl14['Name of the Organization']!='GRAND TOTAL')]
pl15=pl15[(pl15['Name of the Organization']!='GRAND TOTAL')]
pl16=pl16[(pl16['Name of the Organization']!='GRAND TOTAL')]
pl17=pl17[(pl17['Name of the Organization']!='GRAND TOTAL')]
pl18=pl18[(pl18['Name of the Organization']!='GRAND TOTAL')]
pl19=pl19[(pl19['Name of the Organization']!='GRAND TOTAL')]
pl20=pl20[(pl20['Name of the Organization']!='GRAND TOTAL')]
pl21=pl21[(pl21['Name of the Organization']!='GRAND TOTAL')]
```

```python
pl10['Year']=10
pl11['Year']=11
pl12['Year']=12
pl13['Year']=13
pl14['Year']=14
pl15['Year']=15
pl16['Year']=16
pl17['Year']=17
pl18['Year']=18
pl19['Year']=19
pl20['Year']=20
pl21['Year']=21
```

# Salary-wise analysis

```python
PL=pd.concat([pl10,pl11,pl12,pl13,pl14,pl15,pl16,pl17,pl18,pl19,pl20,pl21])
```

```python
intern=PL.loc[PL['Name of the Organization'].str.contains(r"\b(int|intern|internship)\b", flags=re.I, regex=True)]
```

```python
PL=PL.loc[~PL['Name of the Organization'].str.contains(r"\b(int|intern|internship)\b", flags=re.I, regex=True)]
```

```python
PL.fillna(0,inplace=True)
intern.fillna(0,inplace=True)
```

```python
PL
```

```python
PL10=PL[PL['Year']==10][PL.columns.difference(['Year'])]
PL11=PL[PL['Year']==11][PL.columns.difference(['Year'])]
PL12=PL[PL['Year']==12][PL.columns.difference(['Year'])]
PL13=PL[PL['Year']==13][PL.columns.difference(['Year'])]
PL14=PL[PL['Year']==14][PL.columns.difference(['Year'])]
PL15=PL[PL['Year']==15][PL.columns.difference(['Year'])]
PL16=PL[PL['Year']==16][PL.columns.difference(['Year'])]
PL17=PL[PL['Year']==17][PL.columns.difference(['Year'])]
PL18=PL[PL['Year']==18][PL.columns.difference(['Year'])]
PL19=PL[PL['Year']==19][PL.columns.difference(['Year'])]
PL20=PL[PL['Year']==20][PL.columns.difference(['Year'])]
PL21=PL[PL['Year']==21][PL.columns.difference(['Year'])]
```

## OVERALL

```python
PL_all=PL.sort_values(by=['Salary Per Annum (Rs. In Lakhs)','TOTAL'],ascending=False).head(10)
PL_all[['Name of the Organization','Salary Per Annum (Rs. In Lakhs)']]
```

```python
PL_all=PL.sort_values(by=['Salary Per Annum (Rs. In Lakhs)','TOTAL'],ascending=False).head(10)
plt.figure(figsize=(27,15))
plt.title('Top 10 companies in last 10 years by Package offered')
plt.xlabel('Companies')
plt.ylabel('Package offered')
plt.bar(PL_all['Name of the Organization'],PL_all['Salary Per Annum (Rs. In Lakhs)'])

for index, data in enumerate(PL_all['Salary Per Annum (Rs. In Lakhs)']):
    plt.text(index-0.1, data+0.1 , data , fontsize=15)

plt.show()
```

```python
PL_all=PL_all.loc[:, (PL_all != 0).any(axis=0)]
PL_all.set_index('Name of the Organization',inplace=True)
aall=PL_all.drop('Salary Per Annum (Rs. In Lakhs)',axis=1)
aall = aall[ ['TOTAL'] + [ col for col in aall.columns if col != 'TOTAL' ] ]
aall
```

```python
ten=aall.drop('Year',axis=1)
```

```python
colors=['k','m','c','y','r','g','b','darkgray','lightgreen','firebrick']
plt.rcParams['figure.figsize'] = [15, 7]
ten.plot(kind='bar',color=colors,ylabel='No. of recruitments',title='Recruitments for top 10 companies in last 11 years')
plt.show()
```
